### ANALISIS DE TEXTO

Se obtienen dos tipos de documentos como un pequeño escrito sobre Big Data al igual que una recopilacion de comentarios para analisis de texto. Se muestran las diferentes tecnicas de analisis e inspeccion.

```python
f = open('prueba.txt',encoding='utf8')
text = f.read()
print(text)
```

```python
import re
import string

text = text.lower()
text = re.sub('\[.*?¿\]\%', ' ', text)
text = re.sub('\w*\d\w*', '', text)
text = re.sub('[""‘’“”…«»]', '', text)
text = re.sub('\n', ' ', text)

#text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
```

```python
print(text)
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import Counter
```

```python
#TOKENIZACION PALABRAS
tokens_words = word_tokenize(text)

#TOKENIZACION ORACIONES
tokends_sentence = sent_tokenize(text, language='spanish')
```

```python
tokends_sentence[1]
```

```python
#PALABRAS MAS COMUNES
palabras_comunes = Counter(tokens_words)
print('Las palabras mas comunes son. ',palabras_comunes.most_common(10))
```

```python
#MIRAMOS UNA ORACION 
tokends_sentence[0]
```

```python
#TOKENIZACION ELIMINANDO sIGNOS QUE NO SEAN ALPHANUMERICOS y pasando a MINUSCULAS
Tokens = [w for w in word_tokenize(text.lower()) if w.isalpha()]
```

```python
# ESCOGER PALABRAS QUE NO SEAN PALABRAS COMUNES 
from nltk.corpus import stopwords
tokens_stopword = [w for w in Tokens if w not in stopwords.words('spanish')]
```

```python
#AHORA OBSERVAMOS LAS NUEVAS PALABRAS MAS COMUNES CON EL ARREGLO HECHO
New_palabras_comunes = Counter(tokens_stopword)
print(New_palabras_comunes.most_common(10))
```

#### GENSIM

```python
#TOKENIZAMOS CADA ORACION
tokends_word_sentece = [word_tokenize(i) for i in tokends_sentence]
tokends_word_sentece
```

```python
#LIMPIAMOS LA TOKENIZACION
def clean(token):
    for i in token:
        for j in i:
            if ((j.isalpha() == False) or (j in stopwords.words('spanish'))):
                i.remove(j)
```

```python
clean(tokends_word_sentece)
```

```python
from gensim.corpora import Dictionary
dic = Dictionary(tokends_word_sentece)
dic.token2id
```

```python
#TOMAR EL ID DE UNA PALABRA
token_id = dic.token2id.get('textos')
print(token_id)

#TOMAR LA PALABRA DE UN ID
print(dic.get(86))  # get(token_id)
```

```python
#TABLA DE (ID, FRECUENCIA)
corpus = [dic.doc2bow(doc) for doc in tokends_word_sentece]

#VER ESTA TABLA PARA LA TERCERA ORACION (para todas solo corpus)
print(corpus[3])
```

```python
#ORDENAR ESTAS FRECUENCIA
sort_freq = sorted(corpus[3], key=lambda x: x[1], reverse=True)
print(sort_freq)
```

```python
#VEMOS LAS 3 PALABRAS MAS USADAS EN ESTA ORACION
for word_id, word_freq in sort_freq[:5]:
    if(dic.get(word_id) not in stopwords.words('spanish')):
        print(dic.get(word_id), word_freq)
```

```python
#CONTEO TOTAL DE APARICIONES EN TODAS LAS ORACIONES
from itertools import product
import itertools
from collections import defaultdict

#DICCIONARIO VACIO
total_word_count = defaultdict(int)
for word_id, word_count in itertools.chain.from_iterable(corpus):
     total_word_count[word_id] += word_count
        
#YA TENIENDO EL DICCIONARIO DEBEMOS ORGANIZARLO PARA TENER LAS PALABRAS MAS USADAS
sorted_dict_words = sorted(total_word_count.items(), key=lambda x: x[1], reverse=True)        
        
#OBSERVAMOS LAS 3 PALABRAS MAS USADAS EN TODAS LAS CRISTICAS (SIN PALABRAS VACIAS)
for word_id, word_count in sorted_dict_words[:10]:
    if(dic.get(word_id) not in stopwords.words('spanish')):
        print(dic.get(word_id),word_count) 
```

#### TF-IDF
