```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('Train.csv')
df = pd.DataFrame(data)
df_grap = df.copy()
df_grap.head(5)
```

```python
df_grap['Item_Weight'] = df_grap['Item_Weight'].fillna(df_grap.groupby('Item_Type')['Item_Weight'].transform('mean'))
```

```python
df_grap['Outlet_Size'] = df_grap['Outlet_Size'].replace({
    'Medium':1,
    'High': 2,
    'Small' : 0
})

df_grap['Item_Fat_Content'] = df_grap['Item_Fat_Content'].replace({
    'low fat' :'Low Fat',
    'LF': 'Low Fat',
    'reg' : 'Regular',
})
```

```python
df_grap['Outlet_Size'] = df_grap['Outlet_Size'].fillna(df_grap.groupby(['Outlet_Location_Type','Outlet_Type'])['Outlet_Size'].transform('mean'))
df_grap['Outlet_Size'] = df_grap['Outlet_Size'].fillna(df_grap.groupby('Outlet_Type')['Outlet_Size'].transform('mean'))
```

```python
#Tipo de productos en venta 
plt.figure(figsize=(12,4))
g = sns.countplot(y = df_grap['Item_Type'], palette='mako').set(title='TYPE ITEM')
plt.show()
```

Dentro de los productos encontrados en venta para las diferentes tiendas que se encuentran en el dataset se observa una gran variedad de venta para productos relacionados con frutas y vegetales, Snacks seguidos por articulos de casa,comida congelada y lactosa. Los articulos de menor cantidad son los relacionados al desayuno y alimentos de mar

```python
plt.figure(figsize=(16,4))
data = df.groupby('Item_Type')['Item_Visibility'].mean()
fig, ax = plt.subplots(figsize=(16,4))
g = sns.barplot(y = data.values, x = data.index, ax=ax, palette='mako')
g.set(title='ITEM VISIBILITY')
g.set_xticklabels(g.get_xticklabels(), rotation=45)
plt.show()
```

La visibilidad de los productos para cada seccion vienen representados por la siguiente grafica evidenciando una visibilidad levemente alta para los productos de menor cantidad en los supermercados con respecto a los demas como los son articulos de desayuno, comida de mar, los demas mantienen una visibilidad generalizada.

```python
data = df.groupby('Item_Type')['Item_Weight'].mean()

fig, (ax0,ax1) = plt.subplots(ncols=2,nrows=1,figsize=(10,10), sharey=True)
g = sns.barplot(x = data.values, y = data.index, palette='mako', ax=ax0).set(title='ITEM WEIGHT', xlabel='Weight mean', ylabel='Item type')
g = sns.countplot(y = df_grap['Item_Type'], hue=df_grap['Item_Fat_Content'] ,palette='mako', ax=ax1).set(title='FAT CONTENT COUNT', ylabel='Item type')
plt.show()
```

Los articulos con mayor peso se encuentran los enlatados, comida de mar e higiene, aunque son variaciones minimas entre sus pesos. La clasificacion de conteniedo de grasas se encuentra subdividida en baja y regula encontrando que para la mayoria de alimentos su contenido es bajo pero alimentos de las categorias frutas-vegetales, snacks y comida fria tiene un conteniedo regular superior a las demas categorias.

```python
plt.figure(figsize=(16,4))
g = sns.swarmplot(data=df, y='Item_Outlet_Sales', x='Outlet_Identifier').set(title='IDENTIFIER OUTLET SALES', xlabel='Outlet Identifier', ylabel='Sales')
```

Encontramos en el eje x los identificadores de cada una de las tiendas que hacen parte del estudio graficando su relacion con las ventas que estas producen observando que un deficit de ventas para las tiendas OUT010 y OUT019 mientras que las demas mantienen ventas de productos similares con una leve superioridad para la tienda OUT027.

```python
df_grap['Outlet_Size'] = df_grap['Outlet_Size'].replace({
     1:'Medium',
     2: 'High',
     0: 'Small'
})

data = df_grap.groupby(['Outlet_Identifier','Outlet_Location_Type','Outlet_Size','Outlet_Type'])
#data.groups.keys()

cat_Ident = pd.DataFrame(data.groups.keys(), columns=['Identifier','Location Type','Outlet Size','Outlet Type'])
cat_Ident
```

Al agrupar los datos podemos observar que cada identificador de tienda viene relacionado con factores como su locacion, el tamaño de la tienda, el tipo de tienda. Por lo mismo encontramos que las tienda antes mencionadas con pocas ventas OUT010 y OUT019 son de tipo Grocery Store y tamaño pequeño lo cual se refleja claramente en sus ventas de igual manera con OUT027 la cual presento ventas elevadas relacionado con su tipo Supermarket Type3.

```python
sums = df_grap['Item_Outlet_Sales'].groupby(df_grap['Outlet_Type']).sum().reset_index(name='Sales')
plt.figure(figsize=(8,8))
plt.pie(sums['Sales'], labels=sums['Outlet_Type'], rotatelabels=True,autopct='%1.1f%%', shadow=True)
plt.title('SALES FOR OUTLET TYPE')
plt.show()
```

La contribucion para cada tipo de outlet en el total de ventas de productos.

```python
for name in df_grap['Outlet_Identifier'].unique():
  df=df_grap[df_grap['Outlet_Identifier']== name]
  sums = df['Item_Type'].groupby(df['Item_Type']).count().reset_index(name='counts')
  plt.figure(figsize=(10,10))
  plt.title('Outlet_Identifier is'+ name)
  plt.pie(sums['counts'], labels=sums['Item_Type'], rotatelabels=False,autopct='%1.1f%%');
  plt.show()
```

Grafico circular de porcentajes para las categorias ofrecidas detro de cada tienda con su respectivo identificador.
