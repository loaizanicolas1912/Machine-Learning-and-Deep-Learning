```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

datos = pd.read_csv('diabetes2.csv')
df = pd.DataFrame(datos)
df
```

```python
## RELLENAMOS LOS DATOS DE LA GLUCOSA CON EL PROMEDIO DE GLUCOSA DEPENDIENDO DE LA MEDIA DE LOS DIAGNOSTICADOS Y LOS SANOS
df['Glucose'] = df['Glucose'].replace(0,np.nan)
df['Glucose'] = df['Glucose'].fillna(df.groupby('Outcome')['Glucose'].transform('mean'))
df
```

```python
## AHORA OBSERVAMOS LOS DATOS DE LA PRESION SANGINEA
df[df['BloodPressure']==0]
```

```python
## ELIMINAMOS LOS CEROS POR LOS DATOS INTERPOLADOS 
df['BloodPressure'] = df['BloodPressure'].replace(0,np.nan)
df['BloodPressure'] = df['BloodPressure'].interpolate()
```

```python
## ELIMINAMOS LOS CEROS Y LOS REEMPLAZAMOS POR EL DATO ANTERIOR
df['SkinThickness'] = df['SkinThickness'].replace(0,np.nan)
df['SkinThickness'] = df['SkinThickness'].fillna(method='ffill')
```

```python
#PARA LA INSULINA VAMOS A RELLENAROS CON LOS DATOS PROMEDIOS QUE SE TENGAN PARA EL ESTADO DE DIAGNOSTICO
df['Insulin'] = df['Insulin'].replace(0,np.nan)
df['Insulin'] = df['Insulin'].fillna(df.groupby('Outcome')['Insulin'].transform('mean'))
```

```python
## PARA LOS DATOS DE BMI SE PONDRA LA MEDIA
df['BMI'] = df['BMI'].replace(0, df['BMI'].mean())
```

```python
## DEJAMOS LOS VALORES DE GLUCOSA SOLO PARA VALORES ENTEROS IGUAL QUE LA PRESION
df['Glucose'] = df['Glucose'].astype(int)
df['BloodPressure'] = df['BloodPressure'].astype(int)
df['SkinThickness'] = df['SkinThickness'].astype(int)
df
```

```python
## VEMOS QUE EN LAS COLUMNAS DE INSULINA SE ENCUENTRAN MUCHOS DECIMALES
## VAMOS REDONDEAR A UNO 
df['Insulin'] = df['Insulin'].round()
df
```

```python
## VAMOS A AGRUPAR LOS DATOS DE EDAD 
bins = [0,10,20,50,100]
labels = ['Niño','Adolecente','Adulto','Tercera Edad']
df['Age'] = pd.cut(df['Age'], bins=bins, labels=labels)
df
```

```python
### POR ULTIMO VAMOS A GRAFICAR LAS EDADES CON RESPECTO A EL DIAGNOSTICO

fig = plt.figure(figsize=(10,10))
pd.crosstab(df['Age'],df['Outcome']).plot(kind='bar')
plt.xlabel('Edad')
plt.ylabel('Numero de casos por diagnostico')
plt.title('Diagnostico por edad')
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
pd.crosstab(df['Pregnancies'], df['Outcome']).plot(kind='bar')
plt.xlabel('Numero de embarazos')
plt.title('Numeros de embarazos por diagnostico')
plt.show()
```

```python
## VAMOS A ESTANDARIZAR LOS DATOS DEL DATASET PARA QUE TENGAN MEDIA CERO 
## Y DISTRIBUCION ESTANDAR 1

df_Mod = df.iloc[:,0:7]
df_Mod
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_Mod_Scaler = scaler.fit_transform(df_Mod)
```

```python
df_Mod_Scaler = pd.DataFrame(data=df_Mod_Scaler)
df_Mod_Scaler
```

```python
## CREAMOS EL MODELO 
X = df_Mod_Scaler
Y = df['Outcome']

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.20)
model = LogisticRegression(solver='lbfgs',max_iter=1000)
model.fit(X_train,Y_train)
Y_predic = model.predict(X_test)
```

```python
## OBSERVAMOS LOS METRICAS 
model.score(X_test,Y_test)
```

```python
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(Y_predic, Y_test)
print(matriz)
```

```python
from sklearn.metrics import classification_report
report = classification_report(Y_predic, Y_test)
print(report)
```

```python
from sklearn.metrics import roc_auc_score
roc = roc_auc_score(Y_predic,Y_test)
print(roc)
```
