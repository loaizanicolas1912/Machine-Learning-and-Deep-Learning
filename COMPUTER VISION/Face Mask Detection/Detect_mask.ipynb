```python
from keras.models import load_model
from keras.preprocessing import image
from keras.applications.mobilenet_v2 import preprocess_input, decode_predictions
import cv2
import numpy as np

mask_detect = load_model('detec_mask_final.model2')
mask_detect.load_weights('my_weights_final2')
face_cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_frontalface_default.xml')

def face_mask_detector(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray,
                                        scaleFactor=1.2,
                                        minNeighbors=5,
                                        minSize=(110, 110),
                                        flags=cv2.CASCADE_SCALE_IMAGE)
    faces_list=[]
    preds=[]
    for (x, y, w, h) in faces:
        face_frame = frame[y:y+h,x:x+w]
        face_frame = cv2.cvtColor(face_frame, cv2.COLOR_BGR2RGB)
        face_frame = cv2.resize(face_frame, (128, 128))
        face_frame = np.expand_dims(face_frame, axis=0)
        face_frame =  preprocess_input(face_frame)
        final_img = face_frame/255.0
        faces_list.append(face_frame)
        if len(faces_list)>0:
            preds = mask_detect.predict(faces_list)
            print(preds)
        for pred in preds:
            if (preds[0]>0.001):
                Estado = 'Sin Mascarilla'
                cv2.putText(frame,Estado, (100,150), cv2.FONT_HERSHEY_PLAIN,3,(0,0,255),2,cv2.LINE_4)
                cv2.rectangle(frame, (x,y), (x+w,y+h), (0,0,255))
            else:
                Estado = 'Con Mascarilla'
                cv2.putText(frame,Estado, (100,150), cv2.FONT_HERSHEY_PLAIN,3,(0,0,255),2,cv2.LINE_4)
                cv2.rectangle(frame, (x,y), (x+w,y+h), (0,0,255))

    return frame
```

```python
cap = cv2.VideoCapture(0)
ret, frame = cap.read()

while True:
    ret, frame = cap.read()
    output = face_mask_detector(frame)
   

    cv2.imshow('Mask Detection',frame)
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
        
          
cap.release()
cv2.destroyAllWindows()      
```
